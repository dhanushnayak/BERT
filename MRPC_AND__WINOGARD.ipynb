```python

from transformers import AutoTokenizer,TFAutoModelForSequenceClassification
import tensorflow as tf
```

MRPC: takes pair of sentence and determine the closeness  by 
* paraphrase (thoughts and idea)  compare the thoughts by senetence paraphase corpus is used 
* semantic equivalent 

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased-finetuned-mrpc")
model = TFAutoModelForSequenceClassification.from_pretrained("bert-base-cased-finetuned-mrpc")
```

```python
classes  = ['Not','Same']

seq_a  = "A+B^^2 is a formula"
seq_b = "abc def the fucntion to render positive and negative formula"
```

```python
paraphrase = tokenizer.encode_plus(seq_a,seq_b,return_tensors='tf')
```

```python
paraphrase['input_ids'] # takes 2 sequence formes one vector and return yes or no
```

```python
model(paraphrase)
```

```python
logits  = model(paraphrase)[0]
```

```python
tf.nn.softmax(logits,axis=1)#.numpy()
```

```python
result = tf.nn.softmax(logits,axis=1).numpy()[0]
```

**Note** : Index in classes if 0 - No, 1 - Yes, ie. the classes will be ["No","Yes"]

```python
import numpy as np
for i,j in zip(classes,result):
  print(f"Score for {i} = {j}")
```
